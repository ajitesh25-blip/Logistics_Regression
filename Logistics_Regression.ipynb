1. What is Simple Linear Regression (SLR)? Explain its purpose.

Ans. Simple Linear Regression is a statistical method used to model the relationship between one independent variable (predictor) and one dependent variable (response) by fitting a straight line (linear equation) to observed data.The purpose of SLR is to understand how the dependent variable changes as the independent variable changes, and to predict the value of the dependent variable based on a given value of the independent variable.

2. What are the key assumptions of Simple Linear Regression?

Ans. The relationship between the independent variable and the dependent variable is linear, meaning it can be represented by a straight line.The observations are independent of each other, so there is no correlation between the error terms.The variance of the residuals remains constant across all values of the independent variable, which is known as homoscedasticity.The residuals (errors) are normally distributed.

3.  Write the mathematical equation for a simple linear regression model and
explain each term.

Ans. The mathematical equation of a Simple Linear Regression model is given by:
      Y=β0​+β1​X+ε
      In this equation,Y represents the dependent variable, which is the outcome being predicted.X represents the independent variable, which is used to make the prediction. 𝛽0 is the intercept term, which indicates the value of Y when 𝑋=0. β1 is the slope coefficient, which shows the change in Y for a one-unit change in X. ε is the error term, which accounts for the variation in Y that cannot be explained by the linear relationship with X.

5. What is the method of least squares in linear regression?

Ans. The method of least squares is a mathematical technique used in linear regression to estimate the best-fitting line by minimizing the sum of the squares of the differences between the observed values and the predicted values of the dependent variable. It determines the optimal values of the intercept and slope coefficients so that the overall error between the actual data points and the regression line is as small as possible.

6. What is Logistic Regression? How does it differ from Linear Regression?

Ans. Logistic Regression is a statistical method used to model the relationship between one or more independent variables and a categorical dependent variable by estimating probabilities using a logistic (sigmoid) function. Linear Regression is used to predict a continuous dependent variable by fitting a straight line to the data, whereas Logistic Regression is used to predict a categorical outcome (such as yes/no or 0/1) by modeling the probability of occurrence. Linear Regression produces direct numerical values as output, while Logistic Regression outputs probabilities that are transformed into class labels.

7. Name and briefly describe three common evaluation metrics for regression
models.

Ans. Mean Absolute Error (MAE) is a metric that measures the average of the absolute differences between the actual and predicted values, indicating how close the predictions are to the true values without considering direction.Mean Squared Error (MSE) is a metric that calculates the average of the squared differences between the actual and predicted values, giving more weight to larger errors.Root Mean Squared Error (RMSE) is a metric that represents the square root of the mean squared error, providing an error measure in the same units as the dependent variable and making it easier to interpret.

8. What is the purpose of the R-squared metric in regression analysis?

Ans. R-squared is a statistical metric used to measure the proportion of the variance in the dependent variable that is explained by the independent variable(s) in a regression model. Its purpose is to indicate how well the model fits the data, with higher values showing that a greater amount of variation is accounted for by the model.

```python
''' 9. Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.'''

from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 4, 5, 4, 5])

model = LinearRegression()
model.fit(X, Y)

print("Slope (β1):", model.coef_[0])
print("Intercept (β0):", model.intercept_)
```

10.  How do you interpret the coefficients in a simple linear regression model?

Ans. In a simple linear regression model, the slope coefficient (β1) represents the expected change in the dependent variable for a one-unit increase in the independent variable, holding all else constant. The intercept (β0) represents the expected value of the dependent variable when the independent variable is zero. Together, these coefficients define the linear relationship between the independent and dependent variables.
